```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('Cat.png')
print(type(img))
print(img.shape)
plt.imshow(img)
```

```python
#extracting the range of values of the image
print('min value= ',np.min(img),', max value= ',np.max(img))
```

```python
## Selecting all pixels with colours close to black (R=0,G=0,B=0)
threshold=0.5
R=(img[:,:,0]<threshold) & (img[:,:,1]<threshold) & (img[:,:,2]<threshold)

img_black=img.copy()
img_black[~R]=1
img_black[R]=0

plt.imshow(img_black)
```

```python
print(img.ravel().shape)
plt.hist(img[:,:,0].ravel(), color = [1,0,0])
plt.hist(img[:,:,1].ravel(), color = [0,1,0])
plt.hist(img[:,:,2].ravel(), color = [0,0,1])
plt.show()
```

```python
## find the borders of the image

sum_col=np.sum(img_black,axis=0)
w=img.shape
print(w[1])
Ndark=w[0]-sum_col
plt.plot(np.arange(0,w[1]),Ndark)
```

```python
print(np.where(Ndark>50))
```
