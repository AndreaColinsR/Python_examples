# PLOTS !!!

Matplotlib tutorial is [here](https://matplotlib.org/2.0.2/users/pyplot_tutorial.html)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Functionality 

Ideally, a figure should contain:
1. Axes labels including units of measurement (e.g [ms],[s],[$C^o$],etc) 
2. Title
3. If there is more than one line, it should also include a legend

```python
## Your vanilla plot
x=np.array([1,2,3,4,5])
y=np.exp(x)
plt.plot(x,y,'.r')
plt.plot(x,2*y,'.b')
plt.xlabel('Time [s]')
plt.ylabel('Temperature [$C^o$]')
# note that it recognises LaTeX syntax
plt.title('Temperature over time')
plt.legend(['First data', 'Second data'])

## don't forget to show your plot
plt.show()
```

```python
## Subplots

plt.figure(1)
## Each subplot corresponds to a panel within the figure
## divide the figure into 2 rows and 1 column and plot in the first element (here, the first row)
plt.subplot(211)
plt.plot(x,y,'.r')

# Now plot in the second element (second row)
plt.subplot(212)
plt.plot(x,2*y,'.b')
plt.savefig("Subplots.png") #save as png
```

## Aesthetics

 
 - Complete colour list is [here](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
## Colours and Markers
plt.plot(x,y,'-o',linewidth=2.0,color='indigo',alpha=0.3)

## Axes' limits go together in the same line as [xmin, xmax, ymin, ymax]
plt.axis([0, 6, 0, 200])
# save figures
plt.savefig("Aesthetics.png") #save as png
```

## Heatmaps

```python
# join X and Y, but first make sure they're columns
Example_Matrix=np.concatenate((np.expand_dims(x,axis=1),np.expand_dims(y,axis=1)),axis=1)

print(Example_Matrix)
plt.imshow(Example_Matrix,cmap='autumn') 
plt.colorbar()
```
