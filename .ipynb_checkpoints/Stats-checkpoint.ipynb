## Elementary stats with Python


```python
import numpy as np
import matplotlib.pyplot as plt
```

# Mean

```python
x = np.array([[8.0], [1], [2.5], [4], [28.0]])
print(x)
meanx = np.mean(x)
print(meanx)
```

```python
a = np.array([[1, 2], [3, 4],[5,6]])
print(a)
print()
print('The mean of each column is =',np.mean(a,axis=0))
print('The mean of each row is =',np.mean(a,axis=1))
```

## Standard deviation

```python
from scipy import stats 
from numpy.random import seed 
from numpy.random import randn 
from numpy.random import normal 
from scipy.stats import ttest_1samp
# for reproducibility fix the seed
seed(1)
# generate 100 samples of random numbers from a Gaussian distribution of mean=10, std=5
sample1 =normal(10,5,100) 
print('Sample: ', sample1)
print()
print('The mean is =', np.mean(sample1))
print('The std is =', np.std(sample1))


```

```python
plt.hist(sample1, color=[0.5, 0.5, 0.5])
plt.show()
```

### Is the mean of the population equal to zero?

H0=population has mean 0 (null)

H1=population has mean different than 0 (alternative)

```python
alpha=0.05 ## level of acceptance, the lower the alpha, the stricter the test
t_stat, p_value = ttest_1samp(sample1, popmean=0,alternative='two-sided') 
print('T-statistic value:', t_stat) 
print('P-Value: ', p_value)
# If the p-value is <alpha, the test rejects the null hypothesis  
if p_value<alpha:
    print('It\'s not')
else:
    print('It is')
```

# Comparing two independent populations

H0=populations have the same mean (null)

H1=population have different means (alternative)

```python
from scipy.stats import ttest_ind

## Define a second Gaussian population of mean 20, std=5, and 100 samples
sample2 =normal(20,5,100) 

## Plots!
plt.errorbar(1,np.mean(sample1),
            yerr=np.std(sample1),fmt='-o')
plt.errorbar(2,np.mean(sample2),
            yerr=np.std(sample2),fmt='-o')
```

```python
alpha=0.05
t_stat, p_value = ttest_ind(sample1, sample2, alternative='two-sided') 
print('T-statistic value: ', t_stat)  
print('-Value: ', p_value)

# If the p-value is <alpha, the test rejects the null hypothesis  
if p_value<alpha:
    print('It\'s not')
else:
    print('It is')
```

# Correlations

```python
from scipy.stats import pearsonr

plt.scatter(sample1,sample2)
corr, _ = pearsonr(sample1, sample2)
plt.title(['Pearson correlation =%.3f' % corr])
```
