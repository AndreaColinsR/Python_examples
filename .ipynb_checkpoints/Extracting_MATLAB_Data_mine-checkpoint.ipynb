# Converting Russo Data from Matlab to Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
## Load data from Matlab
DataDrake = loadmat('Drake_tt.mat')
print(type(DataDrake))
```

```python
DataDrake.keys()
Pd = DataDrake['Pd']
## for printing PD print(Pd[:,:]), but this looks awful
# but we can print the name of the variables that are inside the structure (this is a numpy array, not a dictionary)
print(Pd.dtype.names)
```

```python
# print a single field instead
xNames=Pd['xNames'][()][0,0]
print(['This variable is a = ', type(xNames)])
print()
print(xNames[0][()][3])
print(xNames.ndim)
print(xNames.shape)
print(xNames.size)
print(xNames.dtype.names)
print()

print(xNames['arr'])
```

```python
#Same for vA (velocity of 4 example sessions)
vA=Pd['vA'][0,0]
print('This variable is a = ', type(vA))
print(vA)
print(vA.ndim)
print(vA.shape)
print(vA.size)
print(vA.dtype)

plt.plot(vA[:,1],vA[:,2])
#plt.plot()
plt.show()
```

```python
#Same for vA (velocity of 4 example sessions)
pA=Pd['pA'][()][0,0]
print('This variable is a = ', type(pA))
print(pA)
print(pA.ndim)
print(pA.shape)
print(pA.size)
print(pA.dtype)

plt.plot(pA[:,1],pA[:,2],'.')
plt.show()
```

```python
mask=Pd['mask'][()][0,0]


print(type(mask))
print(mask.dtype.names)
plt.plot(mask['time'][0][0])
plt.show()
time=mask['time'][0][0]
```

```python
xA=Pd['xA'][()][0,0]
print('This variable is a = ', type(xA))
print(xA.shape)

plt.plot(time,np.mean(xA, axis=1),'.')
plt.show()
```

```python
dir=mask['dir'][0,0]
time=mask['time'][0,0]
dist=mask['dist'][0,0] # total number of cycles in this condition
pos=mask['pos'][0,0]
condNum=mask['condNum'][0,0]
cycleNum=mask['cycleNum'][0,0] # current number of cycle
worldPos=mask['worldPos'][0,0]
```

```python
# choosing especific conditions
cond1=(pos==0) & (dist==7) & (dir==1) & (cycleNum==2)
print(cond1)
```

```python
vAx=vA[:,0:1]
vAy=vA[:,1:2]
#print(vAx[cond1])
print(type(cond1))
```
