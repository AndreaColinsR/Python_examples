# Converting Russo Data from Matlab to Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
## Load data from Matlab
DataDrake = loadmat('Drake_tt.mat')
print(type(DataDrake))
```

```python
DataDrake.keys()
Pd = DataDrake['Pd']
## for printing PD print(Pd[:,:]), but this looks awful
# but we can print the name of the variables that are inside the structure (this is a numpy array, not a dictionary)
print(Pd.dtype.names)
```

```python
#Extracting vA (velocity of 4 example sessions)
vA=Pd['vA'][0,0]
print('This variable is a = ', type(vA))
print(vA)
print(vA.ndim)
print(vA.shape)
print(vA.size)
print(vA.dtype)

plt.plot(vA[:,1],vA[:,2])
#For details on plots, go to Plots.ipynb
plt.show()
```
